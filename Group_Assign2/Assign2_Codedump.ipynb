```python
def parse_county_adjacency(file_path):
    adjacency_dict = {}
    with open(file_path, 'r', encoding='ISO-8859-1') as f:
        current_fips = None
        for line in f:
            parts = line.strip().split('\t')
            if len(parts) > 1 and parts[1]: 
                current_fips = parts[1].zfill(5) 
                adjacency_dict[current_fips] = set()
            if len(parts) > 3 and parts[3]: 
                adjacent_fips = parts[3].zfill(5) 
                adjacency_dict[current_fips].add(adjacent_fips)
    return adjacency_dict


adjacency_dict = parse_county_adjacency(county_adjacency_path)


adjacency_df = pd.DataFrame(
    [(county, adjacent) for county, adjacents in adjacency_dict.items() for adjacent in adjacents],
    columns=['FIPS', 'Adjacent_FIPS']
)

adjacency_df.head()
```

```python


import pulp

model = pulp.LpProblem("Redistricting", pulp.LpMinimize)

num_districts = 10

county_fips_list = merged_df['FIPS'].unique()

decision_variables = {
    (county_fips, district): pulp.LpVariable(f"x_{county_fips}_{district}", cat='Binary')
    for county_fips in county_fips_list
    for district in range(num_districts)
}

total_population = merged_df['TOT_POP'].sum()
ideal_population_per_district = total_population / num_districts

for district in range(num_districts):
    model += (
        pulp.lpSum(merged_df.loc[merged_df['FIPS'] == county_fips, 'TOT_POP'].values[0] *
                   decision_variables[county_fips, district] for county_fips in county_fips_list) 
        >= ideal_population_per_district * 0.9,
        f"Min_Population_District_{district}"
    )
    model += (
        pulp.lpSum(merged_df.loc[merged_df['FIPS'] == county_fips, 'TOT_POP'].values[0] *
                   decision_variables[county_fips, district] for county_fips in county_fips_list) 
        <= ideal_population_per_district * 1.1,
        f"Max_Population_District_{district}"
    )


def add_contiguity_constraints(model, decision_variables, adjacency_df, num_districts):
    for district in range(num_districts):
        for county_fips in county_fips_list:
            adjacent_counties = adjacency_df[adjacency_df['FIPS'] == county_fips]['Adjacent_FIPS'].tolist()
            model += (
                decision_variables[county_fips, district] <=
                pulp.lpSum(decision_variables[adjacent, district] for adjacent in adjacent_counties),
                f"Contiguity_{county_fips}_District_{district}"
            )

add_contiguity_constraints(model, decision_variables, adjacency_df, num_districts)

for county_fips in county_fips_list:
    model += (
        pulp.lpSum(decision_variables[county_fips, district] for district in range(num_districts)) == 1,
        f"One_County_One_District_{county_fips}"
    )

model += pulp.lpSum(
    decision_variables[county_fips, district] * decision_variables[adjacent, district]
    for county_fips in county_fips_list
    for district in range(num_districts)
    for adjacent in adjacency_df[adjacency_df['FIPS'] == county_fips]['Adjacent_FIPS'].tolist()
)

model.solve()

solution_status = pulp.LpStatus[model.status]

if solution_status in ['Optimal', 'Feasible']:
    solution = {
        (county_fips, district): pulp.value(var)
        for (county_fips, district), var in decision_variables.items()
    }
```
